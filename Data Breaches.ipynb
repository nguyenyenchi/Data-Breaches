## Visualise data for deeper data understanding

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Create new Year column
df["Year"] = df["BreachDate"].transform(lambda x: int(x.split("-")[0]))

print(len(df["Year"].unique()), df["Year"].unique())
```

## Review Trends / Seasonality

```python
# Create new column Period using month
df["Period"] = df["BreachDate"].transform(lambda x: int(x.split("-")[1]))

df.head()
```

```python
# Convert BreachDate column from object to date
df["BreachDate"] = pd.to_datetime(df["BreachDate"])
df.dtypes
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(x="BreachDate", y="PwnCount", data=df)
```

```python
# Lineplot for Year=2019

plt.figure(figsize=(12,8))
sns.lineplot(x="BreachDate", y="PwnCount", hue="Year",estimator=None, data=df[df["Year"]==2019])
```
