```python
# Convert string to numpy array and check count of each array in DataClasses column
import numpy as np
df["DataClassesCount"] = df["DataClasses"].apply(lambda x: np.array(eval(x)), 0).apply(lambda x: x.size)
```

```python
df["DataClassesCount"].value_counts()
```

**Observation**

From the above result, we can conclude that most incidents involve the breaches of 4 different data classes.Interestingly, 1 incident involves the breaches of 17 different data classes.

```python
# Check size
print(df[df.DataClassesCount == 17].DataClasses.apply(lambda x: np.array(eval(x)), 0)[643].size)

# Print values
print(df[df.DataClassesCount == 17].DataClasses.apply(lambda x: np.array(eval(x)), 0)[643])
```

```python
# Print first row of DataClasses column
# print(df.loc[0,"DataClasses"])

from ast import literal_eval
# print(df.loc[0,"DataClasses"])
df.DataClasses = df.DataClasses.apply(literal_eval)
```

```python
# Count total occurrences of values in a column of lists
# Use list comprehension with flattening

test = pd.Series([x for item in df.DataClasses for x in item]).value_counts()
print(test)
```

**Observation**

Above result shows that there are 135 data classes in total, among which Email addresses are the most common types to be compromised, followed by passwords, and usernames. 

Few of the types to be compromised the least include password hints, recovery email addresses, and reward program balances.
