```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load the dataset
df = pd.read_csv("databreaches650.csv")

# convert the data into a contingency table with frequencies
crosstab = pd.crosstab(df['IsVerified'], df['IsFabricated'])

# Print the cross-tabulation table
print(crosstab)

# Perform the chi-squared test
chi2, p_value, dof, expected = chi2_contingency(crosstab)
print('Chi-squared statistic:', chi2)
print('P-value:', p_value)
print('Degrees of freedom:', dof)
print('Expected frequencies:', expected)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 

plt.figure(figsize=(12,8))
sns.heatmap(crosstab, annot=True, cmap="YlGnBu")
plt.title('Chi-Square Test Results')
plt.show()
```
